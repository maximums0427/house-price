```python
f_name = 'LotConfig'
def feature_analysis(f_name):
    print('There are {} missing values in this column.'
          .format(train_dataset[f_name].isnull().sum()))
    #print(set(train_dataset['MSZoning']))
    print(test_dataset[f_name].value_counts(normalize=False))
    grps = list(test_dataset[f_name].value_counts(normalize=False).index)
    f_to_plot = []
    for i in grps:
        f_to_plot.append(train_dataset['SalePrice'][train_dataset[f_name]==i])
    fig = plt.figure(1, figsize=(9, 6))

    # Create an axes instance
    ax = fig.add_subplot(111)

    # Create the boxplot
    bp = ax.boxplot(f_to_plot)
    plt.show()
    
    for i in grps:
        train_dataset['SalePrice'][train_dataset[f_name]==i].plot(kind='density',label=i)
    plt.legend()
    plt.show()
feature_analysis(f_name)
```
